```python
import pandas as pd
import numpy as np
import zipfile
from IPython.display import display
import re
```

```python
def extract_data(file, subfolder):
    """"""
    fields = ["id", "nb", "box", "url", "contributors", "title", "sent"]
    list_ = []
    # Open the zipfolder
    zf = zipfile.ZipFile(file, 'r')
    # import all the subfiles into two dataframes
    for field in fields:
        subfile = 'wikipedia-biography-dataset/{}/{}.{}'.format(subfolder, subfolder, field)
        if field == "sent":
            sent = pd.read_csv(zf.open(subfile), sep='\n', header=None, names=[field])
        else:
            df = pd.read_csv(zf.open(subfile), sep='\n', header=None, names=[field])
            list_.append(df)
    zf.close()
    df = pd.concat(list_, axis=1)
    
    # concatenate all sentences together into one
    # create a sentence to id map
    sent_map = df[['id', 'nb']]
    sent_map = sent_map.set_index('id')
    my_ids = sent_map.index.get_level_values('id').unique()
    sent_map = pd.DataFrame([[i] for i in my_ids for n in range(1, sent_map.loc[i, 'nb'].max()+1)], columns = ['id'])
    sentences = pd.concat([sent_map, sent], axis=1)
    sentences = sentences.groupby(['id']).agg(sum).reset_index(drop=False)
    df = df.merge(sentences, on='id', how='left')
    
    return df, sentences
```
